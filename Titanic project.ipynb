# Total Survival Count

```python
titanic.Survived.value_counts()
```

```python
p.figure(figsize=(4,4))
ax=sns.countplot(data=titanic , x="Survived")
ax.bar_label(ax.containers[0])
p.show()
```

# How Passenger Class affects survival

```python
group1= titanic.groupby("Survived").agg({ "Age" : "mean", "Pclass": "mean"})
```

```python
ax=sns.barplot(x="Pclass", y="Survived", data=titanic)
ax.bar_label(ax.containers[0])
p.xlabel("Passenger Class")
p.ylabel("Survived")
p.title("Survival Rate by Passenger class")
```

# Survival rate by gender

```python
ax1=sns.barplot(x="Sex", y="Survived", data=titanic)
ax1.bar_label(ax1.containers[0])
p.xlabel("Gender")
p.ylabel("Survived")
p.title("Survival rate by Gender")
p.show()
```

# Age Distribution of Total Passengers

```python
group1=titanic.loc[(titanic["Age"]<=20)].count()
group2=titanic.loc[(titanic["Age"]>20) & (titanic["Age"]<=40)].count()
group3=titanic.loc[(titanic["Age"]>40) & (titanic["Age"]<=60)].count()
group4=titanic.loc[(titanic["Age"]>60) & (titanic["Age"]<=80)].count()
lbl=["Less than 20", "Between 20 to 40", "Between 40 and 60", "Between 60 and 80"]
mlist=[group1["Age"],group2["Age"],group3["Age"],group4["Age"]]
```

```python
p.pie(mlist, labels=lbl, autopct="%1.1f%%")
p.title("Age")
p.legend()
p.show()
```

# Analysis of Survival Rate on the basis of Age and Fare

```python
sns.relplot(data=titanic, x="Age", y="Fare", col="Survived")
p.show()
```
